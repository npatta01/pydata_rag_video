## OpenAI MultiModal

```python
from llama_index.multi_modal_llms.openai import OpenAIMultiModal
import llama_index.core.multi_modal_llms.generic_utils

from llama_index.core.schema import ImageDocument
from IPython.display import Image
import rich
```

```python
openai_mm_llm = OpenAIMultiModal(
    model="gpt-4o-mini"
)
```

```python
img_bagel_path = "../images/restaurant__absolute_bagels.png"
```

```python
Image(filename=img_bagel_path, width=500, height=500) 
```

```python
img_doc = ImageDocument(image_path =img_bagel_path)
img_doc
```

```python
response_multi = openai_mm_llm.complete(
    prompt="Describe this image",
    image_documents=[img_doc],
)
```

```python
rich.print(response_multi.text)
```
